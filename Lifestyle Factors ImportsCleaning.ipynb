```python
# SMOKE AND BREAST 24
import pandas as pd
bsmo = pd.read_csv("downloads/data_download (24).csv")
bsmo.head()
```

```python
bsmo = bsmo.rename(columns={'value': '% Breast Cancer Who Currently Smoke'})
bsmo.drop(columns='State', axis=1, inplace=True)
bsmo.drop(columns='measure', axis=1, inplace=True)
bsmo.drop(columns='GEOID', axis=1, inplace=True)
bsmo.head()
```

```python
bsmo.to_excel('smokingbreastcancer.xlsx')
```
