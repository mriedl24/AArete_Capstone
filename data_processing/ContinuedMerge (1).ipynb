```python
import pandas as pd
```

```python
# import dataset
dataset = pd.read_excel('PreliminaryDataset.xlsx')
dataset.head()
```

```python
# importing and cleaning obesity sets
obesebc = pd.read_csv('downloads/data_download.csv')
obesebc.head()
```

```python
obesebc.drop('GEOID', axis=1, inplace=True)
obesebc.drop('measure', axis=1, inplace=True)
obesebc.head()
```

```python
obesebc.drop('State', axis=1, inplace=True)
obesebc.head()
```

```python
obesebc = obesebc.rename(columns={'value': '% Breast Cancer Patients Who Are Obese (BMI>30)'})
obesebc.head()
```

```python
merged = pd.merge(dataset, obesebc, on='County', how='inner')
merged.head()
```

```python
merged = merged.drop_duplicates(subset='County', keep='first')
merged.head()
```
