```python
import pandas as pd
df = pd.read_excel('downloads/SemifinalMerge.xlsx')
df.head()
```

# IMPORTING HEALTH FACTOR DATA

```python
# Depression sets
depressionbc = pd.read_excel('downloads/depressionbreastcancer.xlsx')
depressionlc = pd.read_excel('downloads/depressionlungcancer.xlsx')
depressionmel = pd.read_excel('downloads/depressionmelanoma.xlsx')
depressionpro = pd.read_excel('downloads/depressionprostate.xlsx')
```

## Merging depression data to set

```python
merged = pd.merge(df, depressionbc, on='County', how='outer')  
merged.head()
```

```python
merged = pd.merge(merged, depressionlc, on='County', how='outer')
merged.head()
```

```python
merged.drop('Unnamed: 0_x', axis=1, inplace=True)
merged.drop('Unnamed: 0_y', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, depressionmel, on='County', how='outer')
merged.head()
```

```python
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, depressionpro, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

## Merging diabetes data to set

```python
diabetesbc = pd.read_excel('downloads/diabetesbreast.xlsx')
diabeteslung = pd.read_excel('downloads/diabeteslungcorrect.xlsx')
diabetesmel = pd.read_excel('downloads/diabetesmelanoma.xlsx')
diabetespro = pd.read_excel('downloads/diabetesprostate.xlsx')
```

```python
merged = pd.merge(merged, diabetesbc, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, diabeteslung, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, diabetesmel, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, diabetespro, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

## Drop duplicate columns


```python
merged = merged.drop_duplicates(subset='County', keep='first')
```

```python
merged.head()
```

## Merging high blood pressure data to set

```python
bpbreast = pd.read_excel('downloads/breasthighBP.xlsx')
bplung = pd.read_excel('downloads/lunghighBP.xlsx')
bpmel = pd.read_excel('downloads/melanomahighBP.xlsx')
bppro = pd.read_excel('downloads/prostatehighBP.xlsx')
```

```python
merged = pd.merge(merged, bpbreast, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, bplung, on='County', how='outer')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, bpmel, on='County', how='inner')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = pd.merge(merged, bppro, on='County', how='inner')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```

```python
merged = merged.rename(columns={'% Melanoma Patients With High Blood Pressure_y': '% Prostate Cancer Patients With High Blood Pressure_y'})
```

## Merging high cholesterol data to set

```python
highbreast = pd.read_excel('downloads/highchobreast.xlsx')
highlung = pd.read_excel('downloads/highcholung.xlsx')
highmel = pd.read_excel('downloads/highchomelanoma.xlsx')
highpro = pd.read_excel('downloads/highchoprostate.xlsx')
```

```python
merged = pd.merge(merged, highmel, on='County', how='inner')
merged.drop('Unnamed: 0', axis=1, inplace=True)
merged.head()
```
