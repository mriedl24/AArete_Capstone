```python
import pandas as pd
```

```python
cancerdf = pd.read_excel("downloads/Merged_Cancer_Data2.xlsx")
```

```python
cancerdf.head()
```

```python
# Clean the 'County' column
cancerdf['County'] = cancerdf['County'].str.replace(r'\(\d+\)', '', regex=True).str.strip()
cancerdf.head()
```

```python
# Replace 'US (SEER+NPCR)' with 'United States'
cancerdf['County'] = cancerdf['County'].str.replace(r'US\s*\(.*?\)', 'United States', regex=True).str.strip()
cancerdf.head()
```

# Importing education datasets

```python
gradeschool = pd.read_excel("downloads/Education less than 9th.xlsx")
gradeschool.head()
```

```python
gradeschool.drop('FIPS', axis=1, inplace=True)
gradeschool.drop('2023 Rural-Urban Continuum Codes([rural urban note])', axis=1, inplace=True)
gradeschool.drop('Value (Percent)', axis=1, inplace=True)
gradeschool.drop('Rank within US', axis=1, inplace=True)
gradeschool.head()
```

```python
highschool = pd.read_excel("downloads/Education at Least high school (1).xlsx")
highschool.drop('FIPS', axis=1, inplace=True)
highschool.drop('2023 Rural-Urban Continuum Codes([rural urban note])', axis=1, inplace=True)
highschool.drop('Value (Percent)', axis=1, inplace=True)
highschool.drop('Rank within US', axis=1, inplace=True)
highschool.head()
```

```python
college = pd.read_excel("downloads/Education at least bachelors.xlsx")
college.drop('FIPS', axis=1, inplace=True)
college.drop('2023 Rural-Urban Continuum Codes([rural urban note])', axis=1, inplace=True)
college.drop('Value (Percent)', axis=1, inplace=True)
college.drop('Rank within US', axis=1, inplace=True)
college.head()
```

## Merge education data to master set

```python
# Merge the two datasets on the 'County' column
merged = pd.merge(cancerdf, gradeschool, on='County', how='inner')  # Use 'inner' to keep only matching rows
merged.head()
```

```python
merged = pd.merge(merged, highschool, on='County', how='inner')  
merged.head()
```

```python
merged = pd.merge(merged, college, on='County', how='inner') 
merged.head()
```

```python
# Importing insurance dataset
notinsured = pd.read_excel("downloads/Insurance- Percent uninsured in demographic group, people at or below 138% of poverty.xlsx")
notinsured.drop('Rank within US', axis=1, inplace=True)
notinsured.head()
```

## Merge insurance data to master set

```python
merged = pd.merge(merged, notinsured, on='County', how='inner')  
merged.head()
```
